# Ploting in R

## 1) Make a suitable plot indicating legend and title.

```r
w <- sort(rnorm(20)); y1 <- sort(rnorm(20)); y2 <- sort(rnorm(20))
```

```r
# w <- sort(rnorm(10)); y1 <- sort(rnorm(10)); y2 <- sort(rnorm(10))
```

```r
# plot(w, y1)
```

```r
# plot(w, y1, type="l") 
```

```r
plot(w, y1, type="l", xlab = "x values", ylab = "y values", main = "Graph title")
```

```r
# plot(w, y2, lty=2, col="blue")
```

```r
lines(w, y2, lty=2, col="blue")
```

```r
legend(-1.5, 1.5, c("y1", "y2"), lty = c(1, 2), col = c("black", "blue"))
```

## 2) Using w and y1 draw a multiple plots on the same page.

```r
par(mfrow = c(2, 2))
plot(w, y1)
plot(w, y1)
plot(w, y1)
plot(w, y1)
```
